```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris.feature_names
```

```python
iris.keys()
```

```python
target = iris.target_names
```

```python
iris_data = pd.DataFrame(iris.data,columns=['sepal length (cm)','sepal width (cm)',
                                            'petal length (cm)','petal width (cm)'])
target = iris.target
```

```python
iris_data['Target'] = target
```

```python
iris_data.dropna(axis=0,how='any')
```

```python
plt.plot(iris_data['petal width (cm)'],iris_data['Target'],'*')
```

```python
X = iris_data[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
y = iris_data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
train = clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```

```python
target_name = iris.target_names
```

```python
print(classification_report(y_test,predictions,target_names = target_name))
```
